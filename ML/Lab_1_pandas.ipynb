name - имя - неинформативно  
sibsp - кол-во братьев, сестёр, супруга(и) не на борту титаника - неинформативно  
parch - кол-во родителей и детей не на борту титаника - неинформативно  
ticket - номер билета - неинформативно  
embarked - порт посадки - неинформативно

```python
data.drop(['Name', 'SibSp', 'Parch', 'Ticket', 'Embarked'], axis=1, inplace = True)
```

```python
data.Age = 1912 - data.Age
```

```python
data.Age
```

```python
data.rename(columns={'PassengerId':'Id_пассажира', 'Survived': 'Выжил', 'Pclass':'Класс_пассажира', 'Sex': 'Пол', 'Age':'Год_рождения', 'Fare': 'Заработок', 'Cabin': 'Номер_каюты'}, inplace=True)
```

```python
data
```

```python
data.fillna(data.mean().astype(int), inplace=True)
```

```python
data['Номер_каюты'].fillna(value=data['Номер_каюты'].value_counts().index[0].split()[0], inplace=True)
```

```python
data.isna().sum()
```

```python
data.to_csv('data.csv')
```
